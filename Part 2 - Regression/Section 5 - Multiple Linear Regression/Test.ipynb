```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
# Encoding categorical data

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder_x = LabelEncoder()
x[:,3] = labelencoder_x.fit_transform(x[:,3])

# Enter the index number of the categorical column

onehotencoder = OneHotEncoder(categorical_features=[3])
x = onehotencoder.fit_transform(x).toarray()
```

```python
np.set_printoptions(formatter={'float_kind':'{:f}'.format})
x
```

```python
# Avoiding dumy variable trap
x = x[:, 1:]
```

```python
x
```

```python
# Splitting the dataset into the test and training sets

from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
# Fitting the training set

from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Make the prediction

y_pred = regressor.predict(x_test)
```

## Backward Elimination

```python
# Backward Elimination

import statsmodels.formula.api as sm

# According to the formula b0 + b1 * x1 + b2 * x2 + ...
# In LinearRegression we don't need to set b0 but in the state model we have to set
# b0 and that's way we gonna use np.ones() to create an array with 1's and we want that
# array to append at the beginning of the array so that's way the below line was use

x = np.append(arr=np.ones((50, 1)).astype(int), values=x, axis=1)
```

```python
x_opt = x[:, [0,1,2,3,4,5]]
```

```python
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
```

```python
regressor_OLS.summary()
```

```python
# Remove the constant with the highest p values and fit the model without
# the constant
# Repeat the steps until P > SL (SL = 0.05)

x_opt = x[:, [0,1,3,4,5]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt = x[:, [0,3,4,5]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt = x[:, [0,3,5]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```
